### 데이터셋 불러오기

https://www.kaggle.com/datasets/jocelyndumlao/global-fashion-brands/data

JOCELYN DUMLAO가 업로드한 Kaggle Dataset (2001~2021년도 데이터)에 최근 2개년도 수치를 직접 수기로 추가

최종 보고서, 해당 코드파일과 함께 데이터셋 파일 업로드

```python
# 데이터셋 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_excel('Dataset Global Fashion Brands Brand Equity Ranking Growth Rate  COO ROO 2001-2023.xlsx')
dataset.head()
```

### 전체 30개 패션 브랜드 비교분석 그래프

```python
# 브랜드가치 칼럼넘버 (brand equity column number)
brand_equity_columns = [col for col in dataset.columns if 'Equity' in col]

# 브랜드가치 킬럼만 추출 (Filtering for brand equity columns for all brands)
all_brands_data = dataset[['BrandName'] + brand_equity_columns]

# 그래프로 나타내기 쉽게 하기위한 전치 (Transposing the dataset for easier plotting)
transposed_all_brands_data = all_brands_data.set_index('BrandName').T
transposed_all_brands_data.index = range(2001, 2024) # Setting years as index

# 전체 브랜드 가치 변화 추이 그래프로 (Plotting the brand equity trends for all brands)
plt.figure(figsize=(15, 8))
sns.lineplot(data=transposed_all_brands_data, dashes=False)
plt.title('Brand Equity Trends for All Brands (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Brand Equity (USD Billion)')
plt.xticks(range(2001, 2024, 1))
plt.axhline(0, color='black', linestyle='--')
plt.grid(True)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 전체 브랜드 성장률 칼럼 추출 (Filtering the dataset for growth rate columns for all brands)
growth_rate_columns = [col for col in dataset.columns if 'GrowthRate' in col]
all_brands_growth_data = dataset[['BrandName'] + growth_rate_columns]

# 전치과정 (Transposing the dataset for easier plotting)
transposed_all_brands_growth_data = all_brands_growth_data.set_index('BrandName').T
transposed_all_brands_growth_data.index = range(2001, 2024) # Setting years as index

# 전체 브랜드 브랜드 가치 성장률 변화 추이 그래프로 (Plotting the growth rate trends for all brands)
plt.figure(figsize=(15, 8))
sns.lineplot(data=transposed_all_brands_growth_data, dashes=False)
plt.title('Growth Rate Trends for All Brands (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(range(2001, 2024, 1))
plt.axhline(0, color='black', linestyle='--')
plt.grid(True)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 전체 브랜드 순위 칼럼 추출(Filtering the dataset for rank columns for all brands)
rank_columns = [col for col in dataset.columns if 'Rank' in col]
all_brands_rank_data = dataset[['BrandName'] + rank_columns]

# 전치과정 (Transposing the dataset for easier plotting)
transposed_all_brands_rank_data = all_brands_rank_data.set_index('BrandName').T
transposed_all_brands_rank_data.index = range(2001, 2024) # Setting years as index

# 전체 브랜드 순위 추이 그래프로 (Plotting the growth rate trends for all brands)
plt.figure(figsize=(15, 8))
sns.lineplot(data=transposed_all_brands_rank_data, dashes=False)
plt.title('Brand Rank Trends for All Brands (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Rank')
plt.xticks(range(2001, 2024, 1))
plt.yticks(range(0, 101, 5))
plt.gca().invert_yaxis() # 높은 순위일수록 위로 가게끔 y축 반대로돌리기 (higher ranks will be shown higher in the graph)
plt.grid(True)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 브랜드가치 상위 5개 브랜드 비교분석 그래프

```python
# 상위 5개 기업 추출 (top 5 companies based on the latest brand equity values) 
latest_brand_equity_column = brand_equity_columns[-1]
top_5_brands = dataset.nlargest(5, latest_brand_equity_column)['BrandName']

top_5_brand_equity_data = dataset[dataset['BrandName'].isin(top_5_brands)][['BrandName'] + brand_equity_columns]
top_5_growth_data = dataset[dataset['BrandName'].isin(top_5_brands)][['BrandName'] + growth_rate_columns]
top_5_rank_data = dataset[dataset['BrandName'].isin(top_5_brands)][['BrandName'] + rank_columns]

# 앞과 동일, 그래프 표현 위한 행과 열 전치 과정 (Transposing the datasets for easier plotting) 
transposed_top_5_equity_data = top_5_brand_equity_data.set_index('BrandName').T
transposed_top_5_equity_data.index = range(2001, 2024) # 인덱스를 연도로 설정 (years as index) 

transposed_top_5_growth_data = top_5_growth_data.set_index('BrandName').T
transposed_top_5_growth_data.index = range(2001, 2024) 

transposed_top_5_rank_data = top_5_rank_data.set_index('BrandName').T
transposed_top_5_rank_data.index = range(2001, 2024) 

# 상위 5개 브랜드 세 가지 그래프로 (브랜드가치, 브랜드가치 성장률, 순위) (Plotting the brand equity, growth rate, rank trends for the top 5 brands)
plt.figure(figsize=(15, 8))
sns.lineplot(data=transposed_top_5_equity_data, dashes=False)
plt.title('Brand Equity Trends for Top 5 Brands (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Brand Equity (USD Billion)')
plt.xticks(range(2001, 2024, 1))
plt.grid(True)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(15, 8))
sns.lineplot(data=transposed_top_5_growth_data, dashes=False)
plt.title('Growth Rate Trends for Top 5 Brands (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(range(2001, 2024, 1))
plt.axhline(0, color='black', linestyle='--')
plt.grid(True)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(15, 8))
sns.lineplot(data=transposed_top_5_rank_data, dashes=False)
plt.title('Rank Trends for Top 5 Brands (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Rank')
plt.xticks(range(2001, 2024, 1))
plt.yticks(range(0, 101, 5))
plt.gca().invert_yaxis()
plt.grid(True)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### 브랜드 가치 상승/하락 시점 추출
상승/하락율 상위 10개 연도 추출

```python
# 브랜드가치 성장률 키워드로 시점 추출 (Extracting by keyword)
growth_dataset = (
    dataset
    .melt(id_vars=['BrandName'], value_vars=growth_rate_columns, var_name='Year', value_name='GrowthRate')
    )

# 결과 (Displaying the results)
print("Top 10 Highest Growth Rates:")
print(growth_dataset.nlargest(10,'GrowthRate').head(10))

print("\nTop 10 Negative Growth Rates:")
print(growth_dataset.nsmallest(10,'GrowthRate').head(10))
```

```python
# 각 연도간 긍정, 부정의 성장률 개수 산출 (Counting the number of positive and negative growth rates for each year)
positive_growth_counts = growth_dataset.groupby('Year')['GrowthRate'].apply(lambda x: (x > 0).sum())
negative_growth_counts = growth_dataset.groupby('Year')['GrowthRate'].apply(lambda x: (x < 0).sum())

# 표로 한번에 나타내기 (Combining the counts for easier comparison)
growth_counts_combined = pd.DataFrame({'Positive Growth Counts': positive_growth_counts, 
                                       'Negative Growth Counts': negative_growth_counts})

growth_counts_combined.sort_values(by=['Year'], ascending=True)
```

브랜드가치 성장률과 개수를 고려한 상위 3개 연도 (The top 3 best and worst years for the companies, considering both the level of growth rates and their frequency):

### 지역별 비교분석

브랜드 원산지 지역(유럽, 미국)별 비교분석 <br>
각 지역별로 브랜드가치의 평균을 내어 분석에 활용하였음

```python
# 미국, 유럽 브랜드 묶고 추출 (Filtering data for brands from America and Europe)
american_brands_data = dataset[dataset['BrandOriginRegion'] == 'America']
european_brands_data = dataset[dataset['BrandOriginRegion'] == 'Europe']

# 미국, 유럽 브랜드 브랜드가치, 브랜드가치 성장률 합산 후 평균 (Aggregating the average brand equity and growth rate for American and European brands by year)
average_american_equity = american_brands_data[brand_equity_columns].mean()
average_european_equity = european_brands_data[brand_equity_columns].mean()

average_american_growth = american_brands_data[growth_rate_columns].mean()
average_european_growth = european_brands_data[growth_rate_columns].mean()

# 그래프 생성 (Preparing data for visualization)
years = range(2001, 2024)
geo_comparison_data_equity = pd.DataFrame({'American Brands': average_american_equity.values, 
                                                     'European Brands': average_european_equity.values}, index=years)
geo_comparison_data_growth = pd.DataFrame({'American Brands': average_american_growth.values, 
                                                     'European Brands': average_european_growth.values}, index=years)

# 지역별 브랜드가치, 브랜드가치 성장률 비교분석 그래프 (Plotting geographic comparison in brand equity and growth rate trends)
plt.figure(figsize=(15, 8))
sns.lineplot(data=geo_comparison_data_equity)
plt.title('Average Brand Equity Trends by Region (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Average Brand Equity (USD Billion)')
plt.xticks(range(2001, 2024, 1))
plt.grid(True)
plt.legend(title='Region')
plt.show()

plt.figure(figsize=(15, 8))
sns.lineplot(data=geo_comparison_data_growth)
plt.title('Average Growth Rate Trends by Region (2001-2023)')
plt.xlabel('Year')
plt.ylabel('Average Growth Rate (%)')
plt.xticks(range(2001, 2024, 1))
plt.grid(True)
plt.axhline(0, color='black', linestyle='--')
plt.legend(title='Region')
plt.show()
```

### 하위 부문별 비교분석

#### 2023년 데이터 비교

```python
# 2023 데이터 활용: 패션 브랜드 하위 부문별 분포, 평균 브랜드가치, 평균 브랜드가치 성장률 (Extracting 2023 data for brand distribution, average equity, and average growth rate across sub-sectors)
brand_distribution_2023 = dataset['BrandSubSector'].value_counts()

average_equity_2023 = dataset.groupby('BrandSubSector')['Equity2023'].mean()
average_growth_2023 = dataset.groupby('BrandSubSector')['GrowthRate2023'].mean()

# 2023년 한 해 데이터이므로 시계열이 아닌 바 그래프 (Creating dataframes for bar graph visualization)
brand_distribution_2023_df = brand_distribution_2023.reset_index()
brand_distribution_2023_df.columns = ['BrandSubSector', 'Count']

average_equity_2023_df = average_equity_2023.reset_index()
average_growth_2023_df = average_growth_2023.reset_index()

# 그래프 설계 (Plotting the bar graphs for 2023 data)
plt.figure(figsize=(12, 6))
sns.barplot(x='BrandSubSector', y='Count', data=brand_distribution_2023_df)
plt.title('Distribution of Brands Across Sub-sectors (2023)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='BrandSubSector', y='Equity2023', data=average_equity_2023_df)
plt.title('Average Brand Equity Across Sub-sectors (2023)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='BrandSubSector', y='GrowthRate2023', data=average_growth_2023_df)
plt.title('Average Growth Rate Across Sub-sectors (2023)')
plt.xticks(rotation=45)
plt.show()
```

#### 전체 연도 분석

```python
# 각 연도별 브랜드가치 평균 산출 (Calculating the average for each year for brand equity)
numeric_columns_e = ['Equity2001', 'Equity2002', 'Equity2003', 'Equity2004', 'Equity2005', 'Equity2006', 'Equity2007', 'Equity2008', 'Equity2009', 'Equity2010', 'Equity2011', 'Equity2012', 'Equity2013', 'Equity2014', 'Equity2015', 'Equity2016', 'Equity2017', 'Equity2018', 'Equity2019', 'Equity2020', 'Equity2021', 'Equity2022', 'Equity2023']

dataset_copy = dataset.copy()
dataset_copy[numeric_columns_e] = dataset_copy[numeric_columns_e].apply(pd.to_numeric, errors='coerce')

# 하위부문별로 묶기 (Group by 'BrandSubSector' and calculate the mean)
average_growth_by_subset_e = dataset_copy.groupby('BrandSubSector')[numeric_columns_e].mean()

# 결과 (The result)
print(average_growth_by_subset_e)
```

```python
# 브랜드가치 성장률도 동일한 방식으로 (same for growth rates)
numeric_columns_g = ['GrowthRate2001', 'GrowthRate2002', 'GrowthRate2003', 'GrowthRate2004', 'GrowthRate2005', 'GrowthRate2006', 'GrowthRate2007', 'GrowthRate2008', 'GrowthRate2009', 'GrowthRate2010', 'GrowthRate2011', 'GrowthRate2012', 'GrowthRate2013', 'GrowthRate2014', 'GrowthRate2015', 'GrowthRate2016', 'GrowthRate2017', 'GrowthRate2018', 'GrowthRate2019', 'GrowthRate2020', 'GrowthRate2021', 'GrowthRate2022', 'GrowthRate2023']

dataset_copy[numeric_columns_g] = dataset_copy[numeric_columns_g].apply(pd.to_numeric, errors='coerce')

average_growth_by_subset_g = dataset_copy.groupby('BrandSubSector')[numeric_columns_g].mean()

print(average_growth_by_subset_g)
```

```python
# 하위 부문별 브랜드가치 변화 추이 그래프 (Plotting for 'Equity' columns)
average_growth_by_subset_e.T.plot(kind='line', marker='o', figsize=(12, 6), title='Average Equity Growth by Year')
plt.xlabel('Year')
plt.ylabel('Average Equity Growth')
plt.legend(title='BrandSubSector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 하위 부문별 브랜드가치 성장률 변화 추이 그래프
average_growth_by_subset_g.T.plot(kind='line', marker='o', figsize=(12, 6), title='Average GrowthRate by Year')
plt.xlabel('Year')
plt.ylabel('Average GrowthRate')
plt.axhline(0, color='black', linestyle='--')
plt.legend(title='BrandSubSector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
